## Popularity Based Recommendation

```python
df_bwr = df_ratings.merge(df_books, on='ISBN') 
```

```python
df_num_ratings = df_bwr.groupby("Book-Title").count()["Book-Rating"].reset_index()
df_num_ratings = df_num_ratings.rename(columns = {"Book-Rating":"num_ratings"})
df_num_ratings
```

```python
df_bwr.groupby("Book-Title")["Book-Rating"].mean().reset_index()
```

```python
df_avg_ratings = df_bwr.groupby("Book-Title")["Book-Rating"].mean().reset_index()
df_avg_ratings = df_avg_ratings.rename(columns = {"Book-Rating":"avg_rating"})
df_avg_ratings
```

```python
df_popular = df_num_ratings.merge(df_avg_ratings, on = "Book-Title")
df_popular
```

```python
df_popular = df_popular[df_popular["num_ratings"]>250].sort_values('avg_rating', ascending=False).head(50)
```

```python
df_popular = df_popular.merge(df_books, on = "Book-Title").drop_duplicates("Book-Title")
```

```python
df_bwr.head()
```

```python
x = df_bwr.groupby("User-ID").count()["Book-Rating"] > 250
user_250 = x[x].index
user_250
```

```python
filtered_rating = df_bwr[df_bwr["User-ID"].isin(user_250)]
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books = y[y].index
famous_books
```

```python
final_rating = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
final_rating
```

```python
pt = final_rating.pivot_table(index = "Book-Title", columns = 'User-ID', values = 'Book-Rating')
pt.fillna(0, inplace = True)
```

```python
pt
```

## Cosine Similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity

sim_score = cosine_similarity(pt)
```
